````python
from fastText import train_supervised
from fastText import train_unsupervised

model_unsup = train_unsupervised(
    input='./sentences.txt',
    epoch=1,
    model="cbow",
    thread=10,
    minCount=1,
)
````

````python
model_unsup.get_word_vector('らく賃')
````

````python
model_unsup.get_labels(include_freq=True)
````

````python
model_sup = train_supervised(
    input='./sentences.txt',
    epoch=1,
    thread=10,
    minCount=1,
)
````

````python
model_sup.get_word_vector('らく賃')
````

````python
model_sup.get_labels(include_freq=True)
````

````python
model_sup.predict('らく賃 担当者 誰')
````

````python
# 未知語
print(model_sup.predict('金庫 担当者 誰'))
print(model_sup.predict('あなた 担当者 誰'))
print(model_sup.predict('山形県 担当者 誰'))
````

````python
# 未知語のベクトルが取れない
model_sup.get_word_vector('金庫')
````

````python
# 未知語のベクトルが取れない
model_sup.get_word_vector('山形県')
````
