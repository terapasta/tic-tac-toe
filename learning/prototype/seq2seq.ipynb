### 以下、ローカルPCで実行

scpでローカルのPCへファイルを転送する

```
$ scp deploy@app.my-ope.net:/home/deploy/input.txt.tar.gz .
$ scp deploy@app.my-ope.net:/home/deploy/output.txt.tar.gz .
```

scpで機械学習用のPCへファイルを転送する

```
$ scp input.txt.tar.gz harapan@192.168.0.16:/home/harapan
$ scp output.txt.tar.gz harapan@192.168.0.16:/home/harapan
```

機械学習用PCへログインする

```
$ ssh harapan@192.168.0.16
```

### 以下、機械学習用PCで実行

ファイルを規定のディレクトリへコピー

```
$ cd models_nlp/tutorials/rnn/translate
$ cp ~/input.txt.tar.gz data
$ cp ~/output.txt.tar.gz data
$ cd data/
$ tar -zxvf input.txt.tar.gz
$ tar -zxvf output.txt.tar.gz
```

seq2seqにながせる形式に分割

```
$ split -l 20000 input.txt input. 
$ mv input.aa train_in.txt
$ mv input.ab test_in.txt

$ split -l 20000 output.txt output. 
$ mv output.aa train_out.txt
$ mv output.ab test_out.txt
```

seq2seqの学習処理を実行する

```
$ python translate.py --data_dir data --train_dir train --size=256 --num_layers=3

...

global step 9800 learning rate 0.5000 step-time 0.08 perplexity 1.38
  eval: bucket 0 perplexity 145.09
  eval: bucket 1 perplexity 15.75
  eval: bucket 2 perplexity 15.41
  eval: bucket 3 perplexity 9.01
global step 10000 learning rate 0.5000 step-time 0.08 perplexity 1.34
  eval: bucket 0 perplexity 269.44
  eval: bucket 1 perplexity 17.34
  eval: bucket 2 perplexity 7.91
  eval: bucket 3 perplexity 6.68
global step 10200 learning rate 0.5000 step-time 0.09 perplexity 1.37
  eval: bucket 0 perplexity 205.27
  eval: bucket 1 perplexity 15.75
  eval: bucket 2 perplexity 9.27
  eval: bucket 3 perplexity 6.80
```
